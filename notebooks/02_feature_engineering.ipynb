# 02 - Feature Engineering

This notebook shows how to use the reusable feature engineering pipeline from `src/`.

Steps:
- Load a subset of the M5 data.
- Build the full feature set (lags, rolling stats, time, price, events, SNAP).
- Inspect a few rows to verify the features.


```python
from src.config import get_default_config
from src.preprocessing import prepare_base_dataframe
from src.features import build_features
from src.utils import get_logger

logger = get_logger()
cfg = get_default_config()
cfg.subset.subset_type = "state_store_cat"  # or "random_series" for a random subset

base_df = prepare_base_dataframe(cfg.paths, cfg.subset)
feat_df = build_features(base_df)
feat_df.head()
```

```python
feat_cols = [c for c in feat_df.columns if c not in ["id", "item_id", "dept_id", "cat_id", "store_id", "state_id", "d", "date", "sales"]]
print("Number of features:", len(feat_cols))
feat_df[feat_cols].describe().T.head(20)
```
