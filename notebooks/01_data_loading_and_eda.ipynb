# 01 - Data Loading and EDA

This notebook demonstrates how to:

- Load the M5 CSVs using the reusable code in `src/`.
- Apply a small subset for faster exploration.
- Inspect basic shapes and summary statistics.
- Plot a few example time series.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd

from src.config import get_default_config
from src.preprocessing import prepare_base_dataframe
from src.utils import get_logger

logger = get_logger()

cfg = get_default_config()
# For quick EDA, you can also switch to random_series here if desired.
cfg.subset.subset_type = "state_store_cat"

df = prepare_base_dataframe(cfg.paths, cfg.subset)
df.head()
```

```python
print("Number of series:", df["id"].nunique())
print("Number of dates:", df["date"].nunique())
df.describe(include="all").T.head(20)
```

```python
# Plot a few example series
sample_ids = list(df["id"].unique())[:3]
plt.figure(figsize=(12, 8))
for idx, series_id in enumerate(sample_ids, start=1):
    plt.subplot(len(sample_ids), 1, idx)
    sub = df[df["id"] == series_id].sort_values("date")
    plt.plot(sub["date"], sub["sales"])
    plt.title(f"Series: {series_id}")
    plt.ylabel("Sales")

plt.tight_layout()
plt.show()
```
